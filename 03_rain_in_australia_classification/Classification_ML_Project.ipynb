## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
````

````python
df.isnull().sum()
````

````python
# Remove rows where target variable is missing
df.dropna(subset=['RainTomorrow'], inplace=True)
````

````python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['str']).columns
print("Numerical Columns:")
print(num_cols)
print("\nCategorical Columns:")
print(cat_cols)
````

````python
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
````

````python
cat_cols = cat_cols.drop('RainTomorrow')
````

````python
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
````

````python
df.isnull().sum()
````

### What all manipulations have you done and insights you found?


#### Data Manipulations Performed

The following data preprocessing steps were carried out during the data wrangling phase:

- Identified missing values in all features using `isnull().sum()`.
- Removed rows with missing values in the target variable (`RainTomorrow`) because the target variable cannot be imputed in a supervised learning problem.
- Separated the dataset into numerical and categorical features for appropriate preprocessing.
- Imputed missing values in numerical columns using the **median**, as several weather-related variables such as rainfall and evaporation were skewed and contained outliers.
- Imputed missing values in categorical columns using the **mode (most frequent value)** to preserve the existing distribution of categories.
- Verified that all missing values were successfully handled before proceeding to further analysis.

#### Insights

- The dataset contained missing values across multiple numerical and categorical features.
- Features such as **Sunshine**, **Evaporation**, **Cloud9am**, and **Cloud3pm** had a relatively high percentage of missing values (approximately 38–48%). Instead of dropping them immediately, they were retained because these weather-related variables may provide valuable information for predicting rainfall.
- Most of the remaining features had only a small percentage of missing values, making imputation an appropriate choice.
- After preprocessing, the dataset contained no missing values and was ready for exploratory data analysis, feature engineering, and model building.

#### Chart - 1

````python
# Chart - 1 visualization code

plt.figure(figsize=(8, 6))

ax = sns.countplot(
    data=df,
    x='RainTomorrow',
    hue='RainTomorrow',
    palette='viridis',
    legend=False
)

plt.title('Distribution of Rain Tomorrow', fontsize=16, fontweight='bold')
plt.xlabel('Rain Tomorrow', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Add value labels
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}',
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom',
        fontsize=11
    )

plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.show()
````

Understanding the class distribution is important before building classification models, as class imbalance can affect model performance and evaluation.

##### 2. What is/are the insight(s) found from the chart?

- The dataset is **imbalanced**, with significantly more observations where it did not rain the next day (`No`) than where it did (`Yes`).
- Approximately **110,316** records belong to the **No** class, while **31,877** records belong to the **Yes** class.
- Around **77%** of the observations are **No**, and **23%** are **Yes**, indicating that rainy days occur much less frequently than non-rainy days.
- This imbalance suggests that relying only on **accuracy** may give misleading results. Therefore, evaluation metrics such as **Precision, Recall, F1-score, and ROC-AUC** should also be considered during model evaluation.

Reliable rainfall predictions can help organizations optimize planning, reduce operational risks, and allocate resources more effectively.

However, the class imbalance may negatively impact model performance if not handled properly. A model trained without considering this imbalance may become biased toward predicting the majority class (`No`) because it appears more often in the dataset. As a result, the model may fail to detect actual rainy days, leading to missed weather warnings and poor decision-making. Therefore, appropriate evaluation metrics and, if necessary, imbalance-handling techniques should be considered during model development.

````python
# Handling Missing Values & Missing Value Imputation
df_temp = pd.read_csv("weatherAUS.csv")
````

````python
df_temp['Date'] = pd.to_datetime(df_temp['Date'])
df_temp['Month'] = df_temp['Date'].dt.month
````

````python
# Evaporation
df_temp['Evaporation'] = (
    df_temp.groupby(['Location', 'Month'])['Evaporation']
           .transform(lambda x: x.fillna(x.median()))
)

# Sunshine
df_temp['Sunshine'] = (
    df_temp.groupby(['Location', 'Month'])['Sunshine']
           .transform(lambda x: x.fillna(x.median()))
)
````

````python
df_temp['Evaporation'] = df_temp['Evaporation'].fillna(df_temp['Evaporation'].median())
df_temp['Sunshine'] = df_temp['Sunshine'].fillna(df_temp['Sunshine'].median())
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

#### What all missing value imputation techniques have you used and why did you use those techniques?

To handle missing values effectively, different imputation techniques were applied based on the nature of each feature and the percentage of missing values.

**1. Median Imputation (Numerical Features)**

- Applied median imputation to most numerical features with low to moderate missing values.
- The median is less affected by outliers and skewed distributions than the mean, making it a suitable choice for weather-related variables such as rainfall, temperature, pressure, and wind speed.

**2. Mode Imputation (Categorical Features)**

- Applied mode imputation to categorical features such as `WindGustDir`, `WindDir9am`, `WindDir3pm`, and `RainToday`.
- Since categorical variables represent discrete classes, replacing missing values with the most frequently occurring category helps preserve the original distribution of the data.

**3. Location + Month Median Imputation (Evaporation and Sunshine)**

- The `Evaporation` and `Sunshine` features contained a large proportion of missing values (approximately **43%** and **48%**, respectively).
- Instead of using a single global median, missing values were imputed using the **median of observations belonging to the same Location and Month**.
- This approach takes into account both **geographical differences** and **seasonal weather patterns**, resulting in more realistic and meaningful imputations.
- For groups where all values were missing, the overall column median was used as a fallback to ensure that no missing values remained.

**4. Handling Missing Target Values**

- Rows with missing values in the target variable (`RainTomorrow`) were removed because the target variable is essential for supervised learning and cannot be reliably imputed.

**Why were these techniques chosen?**

- Median imputation provides a robust solution for numerical variables with skewed distributions and outliers.
- Mode imputation is the most appropriate technique for categorical variables.
- Location + Month median imputation preserves regional and seasonal weather characteristics, making it more effective than using a single global median for `Evaporation` and `Sunshine`.
- Removing records with missing target values ensures the quality and reliability of the classification model.
